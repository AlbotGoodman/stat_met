## Sannolikhet och uppräkning (kombinatorik)  

Sannolikhet representeras av ett värde mellan 0 och 1. Detta kan även ses som ett procenttal; 0.63 är alltså 63 % sannolikhet.  

1: händer alltid  
0: händer aldrig  

Två sätt att se på sannolikhet: frekventistisk och XXX?  

### Relativ frekvens  
Antalet gånger A inträffade / totala antalet observationer  
P[A] = f / n  

Exempel:  

Vid en kraftanläggning mäts det att under 80 av de närmsta 100 dagarna, slumpmässigt valda från loggarna, nådde kraftförsörjningen ett maximum mellan 18-19. Vad är sannolikheten att max sker mellan 18-19 en annan dag?  

f / n = 80 / 100  

### Klassiska formeln:  
Antalet sätt A kan inträffa / antalet utfall experimentet kan ha (storleken på utfallsrummet)  
P[A] = n(A) / n(S)  

S = {(brown, blue), (blue, brown), (blue, blue), (brown, brown)}  

Notera att genen för blå ögon är recessiv. Alltså kan barnets ögon bara bli blåa vid (blue, blue).  

Sannolikhet för blåa ögon: P = 1 / 4 = 0.25

### Distribution  

För en tärning:  
1+2+3+4+5+6=21  
21 / 6 = 3.5 (medlet)  

Detta kallas väntevärde i statistik. E[X] = µ  
E - expectation  
µ - väntevärde/medel/mean  

X är här en slumpvariabel som antar 1-6 med lika (uniform) sannolikhet.  

Väntevärde för X (med utfall x):  
$ E[X] = ∑_x xf(x) $  

Exempel:  
En hjärtmedicin för efterbehandling vid hjärtinfarkt undersöks. Låt X vara antalet hjärtslag per minut och patient.  

| x | 40 | 60 | 68 | 70 | 72 | 80 | 100 |  
|---|  
| f(x) | 0.01 | 0.04 | 0.05 | 0.80 | 0.05 | 0.04 | 0.01 |  

E[x] = 40*0.01 + 60*0.04 + ... 100*0.01 = 70  

Väntevärdet är ett av flera viktiga kvantitativa mått. De kallas också moment (till en distribution).  

### Varians  

Ett annat är varians.  

Låt X vara en SV med medel µ. Variansen av X, VarX, eller sigma2, ges av:  

$ VarX = \sigma^2 $ 
$ = E[(X-µ)^2]$  
$ = E[x^2] - (E[x])^2 - formel för beräkning$  

Exempel (medicinen sedan tidigare):  

$ E[x^2] = \sum_x x^2f(x) = 40^2*(0.01) + 60^2*(0.04) + ... + 100^2*(0.01) = 4926.4 $  

$ VarX: E[x^2] - (E[x])^2 = 4926.4 - 70^2 = 26.4 $  

Svårt att tolka! Inga enheter och svårt att tyda.
